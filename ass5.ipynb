```python
#Q3.

Multiple inheritance is a feature in object-oriented programming languages where a class can inherit attributes and methods from more than one parent class. This allows a subclass to inherit functionality from multiple base classes, combining the features of all its parents.

class A:
    def method_A(self):
        print("Method from class A")


class B:
    def method_B(self):
        print("Method from class B")


class C(A, B):
    def method_C(self):
        print("Method from class C")
        
# Creating an instance of class C and calling methods from class A, B, and C
obj_C = C()
obj_C.method_A()
obj_C.method_B()
obj_C.method_C()
```

```python
#Q4.

Getter and setter methods are used to control the access and modification of class attributes (variables). They are also known as accessor and mutator methods, respectively. The getter method is responsible for retrieving the value of an attribute, while the setter method is used to set or modify the value of the attribute, typically with additional validation or processing.
   
    
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    # Getter method for the width attribute
    def get_width(self):
        return self._width

    # Getter method for the height attribute
    def get_height(self):
        return self._height

    # Setter method for the width attribute
    def set_width(self, width):
        if width > 0:
            self._width = width
        else:
            print("Width must be greater than 0.")

    # Setter method for the height attribute
    def set_height(self, height):
        if height > 0:
            self._height = height
        else:
            print("Height must be greater than 0.")

    # Method to calculate the area of the rectangle
    def calculate_area(self):
        return self._width * self._height

# Creating an instance of the Rectangle class
rectangle = Rectangle(5, 3)

# Getting the width and height using the getter methods
print("Width:", rectangle.get_width())   # Output: Width: 5
print("Height:", rectangle.get_height()) # Output: Height: 3

# Using the setter methods to modify the width and height
rectangle.set_width(10)
rectangle.set_height(7)

# Getting the updated width and height
print("Updated Width:", rectangle.get_width())   # Output: Updated Width: 10
print("Updated Height:", rectangle.get_height()) # Output: Updated Height: 7

# Calculating and printing the area of the rectangle
print("Area:", rectangle.calculate_area()) # Output: Area: 70
```

```python
#Q5.
Method overriding in Python is a feature that allows a subclass to provide a specific implementation for a method that is already defined in its superclass. When a method is overridden, the subclass's implementation takes precedence over the superclass's implementation. This allows you to customize the behavior of a method in the subclass without modifying the superclass's code.

class Shape:
    def area(self):
        return 0  # Default implementation, to be overridden by subclasses


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # Method overriding: Override the area() method from the Shape class
    def area(self):
        return self.width * self.height


# Create instances of the classes
shape = Shape()
rectangle = Rectangle(5, 10)

# Call the area() method on both instances
print("Area of the shape:", shape.area())  # Output: Area of the shape: 0
print("Area of the rectangle:", rectangle.area())  # Output: Area of the rectangle: 50

In this example, we have a Shape class with a default implementation of the area() method that returns 0. The Rectangle class inherits from Shape and overrides the area() method to provide a specific implementation for calculating the area of a rectangle. When we call the area() method on the rectangle instance, the overridden method in the Rectangle class is executed, calculating and returning the area of the rectangle (width * height).
```
